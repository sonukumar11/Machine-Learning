# 2. WAP to read , remove the null values from employees.csv file and sort the data in descending order.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('employees.csv')
print(df.head())
df = df.dropna(axis=0)
print(df.isnull().sum())
```

```python
data_new = df.sort_values("Salary",ascending=False)
data_new.head()
```

# 3. WAP for reset the index value and remove index column from reset value on employees.csv file

```python
data_new = data_new.reset_index()
data_new.head()
```

```python
data_new = data_new.drop(["index"],axis=1)
data_new.head()
```

# 4. Draw a Histogram , scatter graph , line graph between salary and bonus from employees.csv file

```python
#Line Graph
import matplotlib.pyplot as plt
Salary_col = data_new.loc[:,"Salary"]
Bonus_col = data_new.loc[:,"Bonus %"]
plt.figure(figsize=(8,5))
plt.title("Line Graph . Salary and Bonus",fontsize=14,color="blue")
plt.xlabel("Salary  X-axis ",fontsize=14,color="orange")
plt.ylabel("Bonus  Y-axis ",fontsize=14,color="green")
plt.grid()
plt.plot(Salary_col,Bonus_col,lw=2,color="red",linestyle="solid",label="Line Graph")
plt.legend("best")
plt.show()
```

```python
#Scatter Graph
plt.figure(figsize=(8,5))
plt.title("Scatter Graph . Salary and Bonus",fontsize=14,color="blue")
plt.xlabel("Salary  X-axis ",fontsize=14,color="orange")
plt.ylabel("Bonus  Y-axis ",fontsize=14,color="green")
plt.grid()
plt.scatter(Salary_col,Bonus_col,c ="green",linewidths = 2,marker ="^", edgecolor ="red",s=200)
plt.legend("best")
plt.show()
```

```python
# Histogram - Salary Column
salary = list(Salary_col)
plt.hist(salary,rwidth=0.7)
plt.title("Histogram . Salary Column",fontsize=14,color="green")
plt.show()
```

```python
# Histogram - Bonus Column
bonus = list(Bonus_col)
plt.hist(bonus,rwidth=0.7)
plt.title("Histogram . Bonus % Column",fontsize=14,color="green")
plt.show()
```
