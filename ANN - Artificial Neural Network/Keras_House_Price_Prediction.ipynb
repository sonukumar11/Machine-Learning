```python
X = df.drop('price',axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test , y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 101)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

#output layer
model.add(Dense(1))

#compiling model
model.compile(optimizer = 'adam',loss='mse')
```

```python
#Here , for each epoch our model will calculate the loss based on validation data
#However, our model only trains on training data not on test data
#i.e weights and bias of the neural network will not change based upon the test data or validation data
#batch_size = serving data in the batch of 128 rows....
model.fit(x=X_train,y=y_train,validation_data =(X_test,y_test),batch_size = 128,epochs = 400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
pred = model.predict(X_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
np.sqrt(mean_squared_error(y_test,pred))
```

```python
mean_absolute_error(y_test,pred)
```

```python
from sklearn.metrics import explained_variance_score
explained_variance_score(y_test,pred)
```
