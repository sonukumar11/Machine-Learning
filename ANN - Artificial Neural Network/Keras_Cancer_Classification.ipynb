```python
X = df.drop('benign_0__mal_1',axis = 1).values
```

```python
y = df['benign_0__mal_1'].values
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#Output Layer - Binary Classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer = 'adam' , loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
#This is a case of overfitting
```

```python
#Using callbacks as earlystop to prevent Overfitting..
model = Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#Output Layer - Binary Classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer = 'adam' , loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
#As we tracking validation loss which needs to be minimized therefore mode-min o/w if not getting use 'auto' 
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
#Here , we have stopped the training after 14 epochs....
```

```python
model_loss_early_stop = pd.DataFrame(model.history.history)
model_loss_early_stop.plot()
```

```python
#Using Dropout layers to prevent Overfitting
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

#Means 50% of the neurons will be randomly turned off during each epoch 
#Randomly turned off -- their weights and bias will not change..
#50% of 30.
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

#50% of 15 will be randomly turned off.
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

#Output Layer - Binary Classification
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer = 'adam' , loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
model_loss_early_stop_dropout = pd.DataFrame(model.history.history)
model_loss_early_stop_dropout.plot()
```

```python
#Binary classification
y_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
